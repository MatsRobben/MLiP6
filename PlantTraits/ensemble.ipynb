```python
import pandas as pd

# Load the dataframes
swin = pd.read_csv('/kaggle/input/planttraits2024-ensemble/swin.csv')
swin_tabular = pd.read_csv('/kaggle/input/planttraits2024-ensemble/swin-tabluar.csv')
swin_tabular_all_data = pd.read_csv('/kaggle/input/planttraits2024-ensemble/swin-tabular-all-data.csv')
swin_tabular_xgboost = pd.read_csv('/kaggle/input/planttraits2024-ensemble/swin-tabular-xgboost.csv')
swin_tabular_more_aug = pd.read_csv('/kaggle/input/planttraits2024-ensemble/swin-tabular-more-aug.csv')
swin_tabular_crop_test = pd.read_csv('/kaggle/input/planttraits2024-ensemble/swin-tabular-crop-test.csv')



# Calculate the average for each column (excluding 'id')
averages = (swin.iloc[:, 1:] + 
            swin_tabular_xgboost.iloc[:, 1:] +
            swin_tabular_all_data.iloc[:, 1:] +
            swin_tabular_more_aug.iloc[:, 1:] + 
            swin_tabular_crop_test.iloc[:, 1:]) / 5

# Combine the averages with the 'id' column from one of the dataframes
submission_df = pd.DataFrame({
    'id': swin_tabular['id'],
    'X4': averages['X4'],
    'X11': averages['X11'],
    'X18': averages['X18'],
    'X50': averages['X50'],
    'X26': averages['X26'],
    'X3112': averages['X3112']
})

# Save the result to a CSV file
submission_df.to_csv('submission.csv', index=False)
```
